## 1.	Open a New Notebook and import the required libraires

```python
  import numpy as np       #used for including multi-dimensional array
  import pandas as pd      #preprocess the data
  pd.set_option('max_columns', None)   #in order to break the limit of columns in data frames
  import matplotlib.pyplot as plt    #to plot the graphs
  import seaborn as sns        #data visualisation, rich in statistical calculations
  plt.style.use('ggplot')      #visualise data statistically
  import datetime
  from wordcloud import WordCloud    #collection or cluster of words decipated in different sizes
```

### 2.	Loading the training & testing Dataset

```python
data = pd.read_csv('/content/sample_data/data.csv')
```

```python
data
```

### 3.	Visualizing the Distribution of Revenue with & without Log

```python
fig, ax = plt.subplots(figsize=(16,6))
plt.subplot(1, 2, 1)
#kde=kernel density estimate
sns.distplot(data['revenue'], kde=False);
plt.title('Distribution of revenue');

plt.subplot(1, 2, 2)
#calculate the natural logarithmic value of x+1 where x belongs to all the input array elements.
sns.distplot(np.log1p(data['revenue']), kde=False);
plt.title('Distribution of log revenue')
```

### 4. Budget VS Revenue

```python
data['log_revenue'] = np.log1p(data['revenue'])
data['log_budget'] = np.log1p(data['budget'])
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
sns.scatterplot(data['budget'], data['revenue'])
plt.title('Revenue vs budget');
plt.subplot(1, 2, 2)
sns.scatterplot(data['log_budget'], data['log_revenue']) 
plt.title('log transfromation of revenue vs budget');
```

### 5.	Impact of Film’s Revenue with or without Homepage

```python
data['homepage']
```

```python
data['has_homepage'] = 0
data[['homepage','has_homepage']]
```

```python
data.loc[data['homepage'].isnull() == False, 'has_homepage'] = 1
data[['homepage','has_homepage']]
```

```python
sns.catplot(x='has_homepage', y='revenue', data=data);
plt.title('Revenue for films with and without a homepage');
```

### 6.	Frequent Words in Movie Titles

```python
plt.figure(figsize=(12, 12))
text =  ' '.join(data['original_title'].values)
wordcloud = WordCloud(max_font_size=None,
                     background_color ='white',
                     width =1200, height =1000).generate(text)
plt.imshow(wordcloud)
plt.title('Top words that are used across movie titles')
plt.axis('off')
plt.show()
```

### 7.	Frequent Words in Movie Overviews

```python
plt.figure(figsize=(12, 12))
text =  ' '.join(data['overview'].fillna('').values)
wordcloud = WordCloud(max_font_size=None,
                     background_color ='white',
                     width =1200, height =1000).generate(text)
plt.imshow(wordcloud)
plt.title('Top word across movie overviews')
plt.axis('off')
plt.show()
```

### 8. Revenue vs Popularity

```python
data['log_revenue'] = np.log1p(data['revenue'])
data['log_popularity'] = np.log1p(data['popularity'])
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
sns.scatterplot(data['popularity'], data['revenue'])
plt.title('Revenue vs popularity');
plt.subplot(1, 2, 2)
sns.scatterplot(data['log_popularity'], data['log_revenue']) 
plt.title('log transfromation of revenue vs popularity');
```

In Revenue vs Popularity Graph:

Graph1: In normal graph, data is mixed up and not visualised properly, In this low popularity movies are getting less revenue.

Graph2: In logarithmic graph, data is not mixed up and visualised properly, In this there are movies which are not popular but get good revenue, and movies which are popular have good revenue.
Graph2 is best than Graph1
